```python
# Importando Pandas
import pandas as pd

```

```python
#importando dados
dados = pd.read_csv("Dados/Crime_Data_from_2020_to_Present[1].csv")
```

```python
dados
```

```python
tamanho = dados.shape
rows , columns  = tamanho
```

```python
print(f"{rows} linhas | {columns} colunas")
```

```python
#Colunas importantes 
data_ocorrencia = pd.DataFrame(dados["date_occurred"])
nome_area = pd.DataFrame(dados["area_name"])
crime = pd.DataFrame(dados["crime_description"])
localizacao = pd.DataFrame(dados["location"])
```

```python
colunas_import = pd.concat([data_ocorrencia,nome_area,crime,localizacao], axis = 1)
```

```python
colunas_import
```

```python
#Areas que ocorreram os Crimes
areas =pd.DataFrame(colunas_import["area_name"].value_counts())
```

```python
areas.columns = ["Areas"]
```

```python
areas
```

```python
#Tipos de Crimes
crimes = pd.DataFrame(colunas_import["crime_description"].value_counts().to_frame())
```

```python
crimes.columns = ["Crimes"]
crimes
```

```python
#Numero de localizacao
localizacao= pd.DataFrame(colunas_import["location"].value_counts())
```
